```python
# Import pandas
import pandas as pd

# Read the file
transport_costs = pd.read_excel('transport_costs.xlsx')

# View the DataFrame
print(transport_costs.shape)
transport_costs.head()
```

```python
transport_1 = transport_costs[['region', 'country', 'port', 'sea freight cost', \
                           'road transport cost per km']]

print(transport_1)
```

```python
# Add new columns and assign values
transport_1['total_sea_freight_cost'] = transport_1['sea freight cost'] * 2 * 4
transport_1['total_road_transport_cost'] = transport_1['road transport cost per km'] * 10000
transport_1['total_export_costs'] = transport_1['total_sea_freight_cost'] + transport_1['total_road_transport_cost']

transport_1
```

```python
# Reorder the columns
transport_final = transport_1[['region', 'country', 'port',
                            'sea freight cost', 'total_sea_freight_cost',
                               'road transport cost per km', 'total_road_transport_cost',
                              'total_export_costs']]

transport_final
```

```python
# Save to CSV
transport_final.to_csv('transport_activity.csv')
```
